```python
import requests
from newsapi import NewsApiClient
import pandas as pd
import numpy as np
import json


tracking_url = ('http://newsapi.org/v2/top-headlines?'
       'country=us&'
       'apiKey=18bcaa50d33243b3b1ea2fb19c509c92')


def top_news():
    
    url = tracking_url
    response = requests.get(url)
    r = response.json()
    r = json.dumps(r)
    df = pd.read_json(r)
    
    return df




if __name__=="__main__":
    
    df = top_news()


```

```python
columns = ['source','author','title','description','url','publishedAt','content']
data = {'source':[],'author':[],'title':[],'description':[],'url':[],'publishedAt':[],'content':[]}

for item in df['articles']:
    for c in columns:
        
        if c == "source":
            data[c].append(item[c]['name'])
            print(item[c]['name'])
        else:
            data[c].append(item[c])
```

```python
data = pd.DataFrame(data)
```
